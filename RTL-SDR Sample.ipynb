## [Pythonライブラリ] PyRtlSdr

### Usage: Extracting complex samples

```python
from rtlsdr import RtlSdr

sdr = RtlSdr()

# configure device
sdr.sample_rate = 2.048e6  # Hz
sdr.center_freq = 95e6     # Hz
sdr.freq_correction = 60   # PPM offset value to correct frequency error from the dongle
sdr.gain = 'auto'

samples = sdr.read_samples(256*1024)
print(samples)
sdr.close()
```

### Usage: Plotting the Power Spectral Density(PSD)

```python
from pylab import *

# use matplotlib to estimate and plot the PSD
psd(samples, NFFT=1024, Fs=sdr.sample_rate/1e6, Fc=sdr.center_freq/1e6)
xlabel('Frequency (MHz)')
ylabel('Relative power (dB)')

show()
```
